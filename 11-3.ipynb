```python
import pandas as pd
import time
```

```python
# 원하는 날짜의 데이터 가져오기 
start = "20190101"
last = "20210101"

# 다음날짜를 쉽게 가져오기 위해서 datetime으로 변환
last = pd.to_datetime(last)
# 하루를 더해준다.
last = last + pd.Timedelta(days=1) 

dfs = []

while True:
    df = pyupbit.get_ohlcv('KRW-BTC', to=last)
    last = df.index[0] # 제일처음인덱스값으로 다시 조회
    
    dfs.append(df)
    
    if start in df.index:
        break
    time.sleep(0.1) # api 사용한도로 인해서 지연을 줌.
dfs
```

```python
result = pd.concat(dfs).sort_index() # 4개묶여있던 데이터를 하나의 데이터로 변환, 인덱스 기준으로 정렬 해준다.
print(result) # 불필요한 2018년 데이터도 포함되어있다.
print(result.loc[start: ])
```

```python
def ohlcv(ticker, start, last):
    # 다음날짜를 쉽게 가져오기 위해서 datetime으로 변환
    last = pd.to_datetime(last)
    # 하루를 더해준다.
    last = last + pd.Timedelta(days=1) 

    dfs = []

    while True:
        df = pyupbit.get_ohlcv(ticker, to=last)
        last = df.index[0] # 제일처음인덱스값으로 다시 조회

        dfs.append(df)

        if start in df.index:
            break
        time.sleep(0.1) # api 사용한도로 인해서 지연을 줌.
    result = pd.concat(dfs).sort_index() # 4개묶여있던 데이터를 하나의 데이터로 변환, 인덱스 기준으로 정렬 해준다.
    return result.loc[start: ]
```
